````python
print("hello world lets start the project")

# Mount the Google Drive to load dataset and save on google etc.
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, glob
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as T
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader
from torchvision.models.detection import FasterRCNN
from torchvision.models.resnet import resnet50
from torchvision.models.detection.backbone_utils import BackboneWithFPN
from torchvision.models.detection.rpn import AnchorGenerator
from torchvision.datasets import CocoDetection
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
from PIL import Image
import matplotlib.pyplot as plt

# Enable the debug-friendly CUDA error tracing which makes debugging easier.Ref how to use 🔗 https://pytorch.org/docs/stable/notes/cuda.html#debugging
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

#This is the SimCLR Augmentations & Dataset inspired by the SimCLR augmentation strategy in [1] and [3]

# Apply two random transformations to simulate positive pairs
class ContrastiveTransform:
    def __init__(self, base_transform):
        self.base_transform = base_transform
    def __call__(self, x):
        return self.base_transform(x), self.base_transform(x)

# Typical SimCLR augmentations
#simclr_aug = T.Compose([
#    T.Resize((224,224)),
#    T.RandomHorizontalFlip(),
#    T.ToTensor()
#])
#More agumentations to get better resutls
simclr_aug = T.Compose([
    T.ColorJitter(brightness=0.2, contrast=0.2),
    T.RandomRotation(10),
    T.Resize((224,224)),
    T.RandomResizedCrop((480, 640), scale=(0.8, 1.0)),
    T.RandomHorizontalFlip(),
    T.ToTensor()
])

# FLIR dataset loader for SimCLR we just take the unlabeled images from the thermal train data section
class FLIRContrastiveDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_paths = glob.glob(os.path.join(image_dir, "*.jpg"))
        self.transform = transform
    def __len__(self): return len(self.image_paths)
    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB") #added conversion to the RGB as I was having diffeicutly with grayscale etc
        return self.transform(img)

# Load the data
unlabeled_dir = "/content/drive/MyDrive/AI Masters /Colab Notebooks/FLIR_ADAS_v2/images_thermal_train/data"
dataset_unlabeled = FLIRContrastiveDataset(unlabeled_dir, ContrastiveTransform(simclr_aug))
loader_unlabeled = DataLoader(dataset_unlabeled, batch_size=16, shuffle=True, drop_last=True)
print(f"Loaded {len(dataset_unlabeled)} unlabeled FLIR images for SimCLR pretraining check point")

# 3. SimCLR Model      model architecture based on [1], implemented in [3]       

# Load base ResNet-50 without classifier head, This design is exactly from hhe SimCLR Paper: https://arxiv.org/abs/2002.05709
# They describe removing the classification layer and instead attaching a projection head for contrastive learning.
# In SimCLR, we do not use classification labels. Instead, we extract feature embeddings
# and give them into a projection head with contrastive loss.
# Replacing fc with nn.Identity() ensures we keep only the feature extractor.
base_resnet = models.resnet50(pretrained=False)
base_resnet.fc = nn.Identity()  # Remove classifier "Replacing fc with nn.Identity() just means just return the input features.”
base_resnet = base_resnet.float().to(device)

# SimCLR wrapper with projection head (2-layer MLP) 
# SimCLR takes features from a backbone.
# then adds a projection head that maps them
# into a lower-dimensional space for contrastive learning.
class SimCLRResNet50(nn.Module):
    def __init__(self, base_encoder, feature_dim=128):
        super().__init__()
        self.encoder = base_encoder
        self.projector = nn.Sequential(
            nn.Linear(2048, 512), # First fully-connected layer
            nn.ReLU(),             # Non-linearity
            nn.Linear(512, feature_dim) # Output the embedding 
        )
    def forward(self, x):
        h = self.encoder(x)  # base feature extractor
        z = self.projector(h)  # projection head output
        return h, z
        # - h featues  an be used for the downstream tasks later 
        # - z projected features are used for SSL training with contrastive loss.

model = SimCLRResNet50(base_resnet).to(device)# ResNet-50 backbone + 2-layer projection head.

#NT-Xent Loss (SimCLR loss) Normalized Temperature-Scaled Cross Entropy Loss from [1] 
def nt_xent_loss(z1, z2, temperature=0.5):
    z1, z2 = F.normalize(z1, dim=1), F.normalize(z2, dim=1) # Cosine similarity = dot product of the normalized vectors 
    N = z1.size(0)  # The Batch size (number of positive pairs).
    z = torch.cat([z1, z2], dim=0) # Concatenate the two augmented views so now we have 2N embeddings in one tensor.
    sim = torch.matmul(z, z.T) / temperature   # Compute similarity matrix (2N x 2N) the cosine similarity between z[i] and z[j].
    mask = torch.eye(2*N, dtype=torch.bool).to(z.device) #mask to remove similarity with itself 
    sim.masked_fill_(mask, -9e15) 
    positives = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)], dim=0) # Extract the positive pairs: each sample's two augmentations.
    nom = torch.exp(positives)
    denom = torch.sum(torch.exp(sim), dim=1)
    return -torch.log(nom / denom).mean() ## NT-Xent loss = -log( nom / denom ) averaged over all the samples.

# This section is the optimizer, number of epochs,
# and the checkpoint logic,pretraining SimCLR.
# The checkpoint saving/loading pattern
# common PyTorch practice and the SimCLR reference implementation Paper: https://arxiv.org/abs/2002.05709



optimizer = torch.optim.Adam(model.parameters(), lr=3e-4, weight_decay=1e-5)# Adam optimizer "robust for SSL" with learning rate 3e-4 and small weight decay. The weight decay helps prevent overfitting
num_epochs_pretrain = 20 #anyting over 20 was taken too long and had runntime disconnect issues etc.The other Paper: https://arxiv.org/abs/2002.05709 had 800 not practical for me
loss_values = []
checkpoint_name = "/content/drive/MyDrive/AI Masters /Colab Notebooks/simclr_checkpoint.pth"
start_epoch = 0 

# Load checkpoint if available
if os.path.exists(checkpoint_name):
    print("Resuming SimCLR from the checkpoint...")
    checkpoint = torch.load(checkpoint_name, map_location=device)
    model.load_state_dict(checkpoint["model"])
    optimizer.load_state_dict(checkpoint["optimizer"])
    loss_values = checkpoint["loss_values"]
    start_epoch = checkpoint["epoch"] + 1

# Training loop 
for epoch in range(start_epoch, num_epochs_pretrain):
    model.train()
    total_loss = 0
    for img1, img2 in loader_unlabeled:
        img1, img2 = img1.to(device), img2.to(device)
        _, z1 = model(img1)
        _, z2 = model(img2)
        loss = nt_xent_loss(z1, z2)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(loader_unlabeled)
    loss_values.append(avg_loss)
    print(f"[SimCLR] Epoch {epoch+1}/{num_epochs_pretrain}, Loss: {avg_loss:.4f}")

    # Save the checkpoint incase I lose the connection with Colab etc
    torch.save({
        "epoch": epoch,
        "model": model.state_dict(),
        "optimizer": optimizer.state_dict(),
        "loss_values": loss_values
    }, checkpoint_name)

# Save the encoder for downstream the Faster R-CNN use
torch.save(model.encoder.state_dict(), "/content/drive/MyDrive/AI Masters /Colab Notebooks/simclr_resnet50_backbone.pth")

# Plot loss the curve for the write up
plt.plot(range(1, len(loss_values)+1), loss_values, marker='o')
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.title("SimCLR Pretraining Loss")
plt.show()



# FLIR Dataset with COCO Labels

from torchvision.datasets import CocoDetection

train_root = "/content/drive/MyDrive/AI Masters /Colab Notebooks/FLIR_ADAS_v2/images_thermal_train"
val_root = "/content/drive/MyDrive/AI Masters /Colab Notebooks/FLIR_ADAS_v2/images_thermal_val"
test_root = "/content/drive/MyDrive/AI Masters /Colab Notebooks/FLIR_ADAS_v2/video_thermal_test"

train_ann = train_root + "/coco.json"
val_ann = val_root + "/coco.json"
test_ann = test_root + "/coco.json"



# Load category IDs globally from all splits =======add where I get this
train_coco = COCO(train_ann)
val_coco = COCO(val_ann)
test_coco = COCO(test_ann)

all_cats = set([a["category_id"] for a in train_coco.anns.values()] +
               [a["category_id"] for a in val_coco.anns.values()] +
               [a["category_id"] for a in test_coco.anns.values()])
cat_ids_global = sorted(list(all_cats))
cat2label_global = {cid: i+1 for i, cid in enumerate(cat_ids_global)}

# Custom COCO wrapper to format outputs ================================================add where I got this and why
class CocoDetectionWrapper(CocoDetection):
    def __init__(self, root, annFile, transform=None, cat_ids=None, cat2label=None):
        super().__init__(root, annFile, transform)
        self.cat_ids = cat_ids
        self.cat2label = cat2label
    def __getitem__(self, idx):
        img, target = super().__getitem__(idx)
        img = img.convert("RGB")
        boxes, labels = [], []
        for obj in target:
            cid = obj['category_id']
            if cid not in self.cat2label: continue
            xmin, ymin, w, h = obj['bbox']
            if w <= 0 or h <= 0: continue
            boxes.append([xmin, ymin, xmin + w, ymin + h])
            labels.append(self.cat2label[cid])
        if len(boxes) == 0:
            return self.__getitem__((idx + 1) % len(self))
        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        return T.ToTensor()(img), {"boxes": boxes, "labels": labels, "image_id": torch.tensor([idx])}

def collate_fn(batch): return tuple(zip(*batch))

train_dataset = CocoDetectionWrapper(train_root, train_ann, cat_ids=cat_ids_global, cat2label=cat2label_global)
val_dataset = CocoDetectionWrapper(val_root, val_ann, cat_ids=cat_ids_global, cat2label=cat2label_global)
test_dataset = CocoDetectionWrapper(test_root, test_ann, cat_ids=cat_ids_global, cat2label=cat2label_global)

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False, collate_fn=collate_fn)

print(f"Loaded {len(train_dataset)} training, {len(val_dataset)} validation, {len(test_dataset)} test images")


# Now Build Faster R-CNN with SimCLR and the Supervised Backbones


from torchvision.models.detection import FasterRCNN
from torchvision.models.detection.backbone_utils import BackboneWithFPN
from torchvision.models.detection.rpn import AnchorGenerator

return_layers = {"layer1": "0", "layer2": "1", "layer3": "2", "layer4": "3"} #why do this =========================================================
in_channels_list = [256, 512, 1024, 2048]
out_channels = 256
anchor_generator = AnchorGenerator(
    sizes=((16,), (32,), (64,), (128,), (256,)),
    aspect_ratios=((0.5, 1.0, 2.0),) * 5
)
num_classes = len(cat_ids_global) + 1

# Get SSL backbone from SimCLR ==============why do this
resnet_ssl = resnet50(weights=None)
resnet_ssl.fc = nn.Identity(); resnet_ssl.avgpool = nn.Identity()
state_dict = torch.load("/content/drive/MyDrive/AI Masters /Colab Notebooks/simclr_resnet50_backbone.pth", map_location="cpu")
resnet_ssl.load_state_dict(state_dict, strict=False)
backbone_ssl = BackboneWithFPN(resnet_ssl, return_layers, in_channels_list, out_channels)
backbone_ssl.out_channels = out_channels

faster_rcnn_ssl = FasterRCNN(
    backbone_ssl,
    num_classes=num_classes,
    rpn_anchor_generator=anchor_generator,
    image_mean=[0.5,0.5,0.5],
    image_std=[0.25,0.25,0.25],
    min_size=512,
    max_size=1024
).to(device)

# The Supervised ResNet backbone
resnet_sup = resnet50(weights="IMAGENET1K_V1") # why do this and what does it mean -------------------------------------------------------------------------
resnet_sup.fc = nn.Identity(); resnet_sup.avgpool = nn.Identity()
backbone_sup = BackboneWithFPN(resnet_sup, return_layers, in_channels_list, out_channels)
backbone_sup.out_channels = out_channels

faster_rcnn_supervised = FasterRCNN(
    backbone_sup,
    num_classes=num_classes,
    rpn_anchor_generator=anchor_generator,
    image_mean=[0.5,0.5,0.5],
    image_std=[0.25,0.25,0.25],
    min_size=512,
    max_size=1024
).to(device)

print("The SSL and Supervised Faster R-CNN baselines are build debug out")


# The COCO Evaluation Function to measure the performace etc =====================================================================

def coco_evaluate(model, dataset, ann_file, device, max_samples= 3500):
    model.eval(); coco_gt = COCO(ann_file); results = []
    with torch.no_grad():
        for idx in range(min(len(dataset), max_samples)):
            img, _ = dataset[idx]
            img_id = dataset.ids[idx]
            pred = model([img.to(device)])[0]
            for box, score, label in zip(pred["boxes"].cpu(), pred["scores"].cpu(), pred["labels"].cpu()):
                if label < 1 or label > len(cat_ids_global): continue
                x1, y1, x2, y2 = box.tolist()
                results.append({
                    "image_id": img_id,
                    "category_id": cat_ids_global[label-1],
                    "bbox": [x1, y1, x2-x1, y2-y1],
                    "score": float(score)
                })
    if not results:
        print("There were no predictions made") #debug just incase added here.
        return None
    coco_dt = coco_gt.loadRes(results)
    coco_eval = COCOeval(coco_gt, coco_dt, "bbox")
    coco_eval.evaluate(); coco_eval.accumulate(); coco_eval.summarize()
    return coco_eval


# The training Loop (Shared) for all

def train_model(model, optimizer, train_loader, val_dataset, val_ann, checkpoint_name, best_model_path, label):
    best_map = 0.0
    train_loss_history, val_map_history = [], []
    start_epoch = 0
    num_epochs = 20

    if os.path.exists(checkpoint_name): #check point trick from https://github.com/sthalles/SimCLR/blob/master/simclr.py
        print("I am Resuming {label} from checkpoint...")
        checkpoint = torch.load(checkpoint_name, map_location=device,weights_only=False) #set the weights to flase as i was getting errors I'm only loading state dic
        model.load_state_dict(checkpoint["model"])
        optimizer.load_state_dict(checkpoint["optimizer"])
        train_loss_history = checkpoint["train_loss_history"]
        val_map_history = checkpoint["val_map_history"]
        best_map = checkpoint["best_map"]
        start_epoch = checkpoint["epoch"] + 1

    for epoch in range(start_epoch, num_epochs):
        model.train()
        running_loss = 0.0
        for images, targets in train_loader:
            images = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
            loss_dict = model(images, targets)
            losses = sum(loss for loss in loss_dict.values())
            optimizer.zero_grad(); losses.backward(); optimizer.step()
            running_loss += losses.item()

        avg_loss = running_loss / len(train_loader)
        train_loss_history.append(avg_loss)
        print(f"[{label}] Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")
        plt.plot(train_loss_history)

#Plot the loss curves for the write up
        plt.plot(train_loss_history)
        plt.title(f'{label} Training Loss')
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.show()

        coco_eval = coco_evaluate(model, val_dataset, val_ann, device)
        if coco_eval:
            val_map = coco_eval.stats[0]
            val_map_history.append(val_map)
            print(f"Epoch {epoch+1} {label} Validation mAP: {val_map:.4f}")
            if val_map > best_map:
                best_map = val_map
                torch.save(model.state_dict(), best_model_path)
                print(f"The New best {label} model (mAP={best_map:.4f}")

        torch.save({
            "epoch": epoch,
            "model": model.state_dict(),
            "optimizer": optimizer.state_dict(),
            "train_loss_history": train_loss_history,
            "val_map_history": val_map_history,
            "best_map": best_map
        }, checkpoint_name)

# Train/Evaluate the Models section  -----------------------------------------------------where and how

optimizer_ssl = torch.optim.SGD(faster_rcnn_ssl.parameters(), lr=0.005, momentum=0.9, weight_decay=0.0005)#why do we optimise the SSL ????????????????????????
train_model(
    faster_rcnn_ssl, optimizer_ssl, train_loader, val_dataset, val_ann,
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_ssl_checkpoint.pth",
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_resnet50_simclr_best.pth",
    "SSL Faster R-CNN"
)

optimizer_sup = torch.optim.SGD(faster_rcnn_supervised.parameters(), lr=0.005, momentum=0.9, weight_decay=0.0005)
train_model(
    faster_rcnn_supervised, optimizer_sup, train_loader, val_dataset, val_ann,
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_supervised_checkpoint.pth",
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_resnet50_supervised_best.pth",
    "Supervised Faster R-CNN"
)


# The Final Test Evaluation on both of the models with coco on the test data

print("\n This is the final Test Evaluation (SSL Faster R-CNN)")
faster_rcnn_ssl.load_state_dict(torch.load(
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_resnet50_simclr_best.pth"))
coco_evaluate(faster_rcnn_ssl, test_dataset, test_ann, device)

print("\n This is the final Test Evaluation (Supervised Faster R-CNN)")
faster_rcnn_supervised.load_state_dict(torch.load(
    "/content/drive/MyDrive/AI Masters /Colab Notebooks/faster_rcnn_resnet50_supervised_best.pth"))
coco_evaluate(faster_rcnn_supervised, test_dataset, test_ann, device)
````
